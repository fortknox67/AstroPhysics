```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=np.arange(0,6.2831853,0.1)
```

```python
def OrbitalDistance(T,e):

    return ((1-(e**2))/(1+e*(np.cos(T))))
```

start with our distance formula 
$r = (1-e^{2})/(1+e*cos(\theta))$
we know that
$(1-e^2)= L^2/(GMm^2)=(r^2v^2)/(GM)$
We then plug in and solve for v to get
$v = (GM(1-e^2))/(r^2)$

We could use our original equation for r and solve, or we could just plug it in in python... which is probably easier.

Sadly this has a mass in it, but I see no way to effectively get rid of it without introducing other variables, so let's just assume it is M=1 for now (which is wildly inaccurate, but it'll be fine.)

```python
G=6.67408*10**(-11)
```

```python
def OrbitalVelocity(T,e):

    return ((G*(1-e**2))/(OrbitalDistance(T,e)))
```

```python
e2 = plt.subplot(111,projection = 'polar')
e2.plot(OrbitalVelocity(T,0.0))
e2.text(1,0,"orbital speed (m/s)")
```

```python
e2 = plt.subplot(111,projection = 'polar')
e2.plot(OrbitalVelocity(T,0.5))
e2.text(1,0,"orbital speed (m/s)")
```
